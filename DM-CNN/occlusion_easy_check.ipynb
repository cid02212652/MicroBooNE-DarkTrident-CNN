```python
import pandas as pd

scores_path = "/home/hep/an1522/dark_tridents_wspace/outputs/inference/run1_signal_resnet34_gn/run1_dt_ratio_0.6_ma_0.05_pi0_larcv_cropped_scores.csv"

df = pd.read_csv(scores_path)

# Make sure entry_number is an int (yours look like floats in the CSV)
df["entry_number"] = df["entry_number"].astype(int)

df.head()
```

```python
def top_by_score(df, n=20, min_pixels=None):
    x = df
    if min_pixels is not None:
        x = x[x["n_pixels"] >= min_pixels]
    return x.sort_values("signal_score", ascending=False).head(n)[
        ["run_number","subrun_number","event_number","entry_number","signal_score","n_pixels"]
    ]

top_by_score(df, n=10)
# top_by_score(df, n=10, min_pixels=500)
```

```python
import numpy as np

def nearest_score(df, target_score, n=10, min_pixels=None):
    x = df
    if min_pixels is not None:
        x = x[x["n_pixels"] >= min_pixels]
    d = (x["signal_score"] - target_score).abs()
    out = x.loc[d.nsmallest(n).index]
    return out.sort_values("signal_score")[["run_number","subrun_number","event_number","entry_number","signal_score","n_pixels"]]

nearest_score(df, target_score=0.085, n=10)
```

```python
def top_by_pixels(df, n=20, min_score=None):
    x = df
    if min_score is not None:
        x = x[x["signal_score"] >= min_score]
    return x.sort_values("n_pixels", ascending=False).head(n)[
        ["run_number","subrun_number","event_number","entry_number","signal_score","n_pixels"]
    ]

top_by_pixels(df, n=10)
```

```python
def nearest_pixels(df, target_pixels, n=10, min_score=None):
    x = df
    if min_score is not None:
        x = x[x["signal_score"] >= min_score]
    d = (x["n_pixels"] - target_pixels).abs()
    out = x.loc[d.nsmallest(n).index]
    return out.sort_values("n_pixels")[["run_number","subrun_number","event_number","entry_number","signal_score","n_pixels"]]

nearest_pixels(df, target_pixels=2000, n=10)
```

```python
def filter_range(df, score_min=None, score_max=None, pixels_min=None, pixels_max=None):
    x = df
    if score_min is not None:   x = x[x["signal_score"] >= score_min]
    if score_max is not None:   x = x[x["signal_score"] <= score_max]
    if pixels_min is not None:  x = x[x["n_pixels"] >= pixels_min]
    if pixels_max is not None:  x = x[x["n_pixels"] <= pixels_max]
    return x[["run_number","subrun_number","event_number","entry_number","signal_score","n_pixels"]]

filter_range(df, score_min=0.05, pixels_min=1500).head(20)
```

```python
from pathlib import Path

folder = Path("/home/hep/an1522/dark_tridents_wspace/outputs/inference/run1_signal")

dfs = []
for p in sorted(folder.glob("*_scores.csv")):
    tmp = pd.read_csv(p)
    tmp["entry_number"] = tmp["entry_number"].astype(int)
    tmp["source_file"] = p.name
    dfs.append(tmp)

all_df = pd.concat(dfs, ignore_index=True)

print("Loaded rows:", len(all_df), "from", len(dfs), "files")
all_df.head()
```
